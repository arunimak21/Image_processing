Converting RGB to Greyscale using Opencv

```python
#importing Opencv
import cv2
image=cv2.imread('download.jpg')
```

```python
type(image)
```

```python
image.shape
```

```python
grey=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
type(grey)
```

```python
grey.shape
```

```python
plt.imshow(grey,cmap="gray")
plt.show()
```

```python
cv2.imwrite('grey_dog.jpg',grey)
```
